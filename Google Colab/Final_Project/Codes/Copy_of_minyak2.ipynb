```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/2020-2021.xlsx')
```

```python
df
```

```python
df2 = df.copy()
```

```python
df2.shape
df2.dropna()
```

```python
df2[df2.columns[2:]] = df2[df2.columns[2:]].apply(lambda x: x.str.replace('Rp','')).apply(lambda x: x.str.replace(',','')).apply(lambda x: x.str.replace('-',''))
```

```python
df2[df2.columns[2:]] = df2[df2.columns[2:]].apply(lambda x: x.str.strip())
```

```python
df2 = df2.replace(r'^\s*$', np.nan, regex=True)
```

```python
df2
```

```python
df2.dropna(inplace=True)
```

```python
df2[df2.columns[2:]] = df2[df2.columns[2:]].astype(float)
```

```python
df2.dtypes
```

```python
df2.fillna(0)
```

```python
df2.to_csv('clean_2020-2022.csv',index=False)
```
