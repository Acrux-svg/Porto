```python
from transformers import pipeline

pretrained_name = "sahri/indonesiasentiment"

nlp = pipeline(
    "sentiment-analysis",
    model=pretrained_name,
    tokenizer=pretrained_name
)

nlp("tidak jelek tapi keren")
```
