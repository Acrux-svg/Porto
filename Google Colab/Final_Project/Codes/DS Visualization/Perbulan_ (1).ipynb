```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Clean_time_zone.csv')
df 
```

```python
from datetime import datetime
#dt = df['Tanggal']
#dt = dt.strftime('%d')
df['Tanggal'] = pd.to_datetime(df.Tanggal, format='%Y-%m-%d')
df['Tanggal'] = df['Tanggal'].dt.strftime('%Y %b')
df['Tanggal'] 
```

```python
plt.figure(figsize=(7,3))
plt.plot(df['Tanggal'],df['mean_wib'],label = 'WIB')
plt.plot(df['Tanggal'],df['mean_wita'],label = 'WITA')
plt.plot(df['Tanggal'],df['mean_wit'],label = 'WIT')
plt.xticks(rotation=90)
plt.title('DATASET MINYAK GORENG')
plt.xlabel('TANGGAL')
plt.ylabel('HARGA MINYAK GORENG')
plt.legend()
plt.show()
```

Pembahasan Grafik : 
Diagram tersebut menjelaskan kenaikan harga minyak goreng yang terjadi di 3 bagian waktu Indonesia yang meliputi wilayah WIB, WITA dan WIT. Dimana titik koordinat X adalah penunjuk secara horizontal, sedangkan titik koordinat Y merupakan penunjuk secara vertikal. X merupakan tanggal dan Y merupakan harga dari minyak goreng. 
Pada data tersebut mengalami kenaikan dan penurun harga yang cukup drastis. Dimana pada Januari 2020 - Januari 2022 mengalami kenaikan harga yang signifikan. Harga tertinggi mencapai Rp. 27.500 - Rp.28.000 dimana terjadi di daerah WIT. Namun pada Maret 2022 mengalami penurunan harga, dan mengalami kenaikan harga lagi pada April 2022 - Mei 2022. Harga minyak tersebut di tiga bagian waktu mengalami penurun secara bersamaan pada rentang bulan May 2022- Juni 2022.
