```python
df['default '].replace('Y',1,inplace = True)
df['default '].replace('N',0,inplace = True)
```

```python
df.SEX = df.SEX.map({'M' : 1, 'F' : 0})
df['SEX'].value_counts()
```

```python
df.MARRIAGE = df.MARRIAGE.map({'Married' : 1, 'Single' : 0, 'Other' : 0, '0' : 0 })
df['MARRIAGE'].value_counts()
```

```python
df.EDUCATION = df.EDUCATION.map({'University' : 1, 'Graduate school' : 1, 'High School' : 1, 'Unknown' : 0, '0' : 0, 'Others' : 0})
df['EDUCATION'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
number_col = ['LIMIT_BAL', 'AGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6',
           'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
          'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4','PAY_AMT5', 'PAY_AMT6']

scaler = MinMaxScaler()
scaler.fit(df[number_col])
df[number_col] = scaler.transform(df[number_col])

df[number_col]
```

```python
df.head()
```

Split data kedalam Train dan tes set

```python
X = df.iloc[1:30000,1:24].values
y = df.iloc[1:30000,24].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)

lr_prediction = model.predict(X_test)
print('akurasi logistik regresi = ',metrics.accuracy_score(lr_prediction,y_test))
```

```python
print("y_predicted",lr_prediction)
print("y_test",y_test)
```
